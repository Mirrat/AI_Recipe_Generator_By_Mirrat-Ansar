```python
import re

def get_best_matches(df, user_ingredients, top_n=3):
    """
    Returns the top_n recipes from df that contain ALL user_ingredients
    by checking if each user ingredient is a substring of at least one
    recipe ingredient. Sorts by fewest extra ingredients.
    """
    user_ingredients_list = [ing.lower().strip() for ing in user_ingredients]
    
    matches = []
    for _, row in df.iterrows():
        # recipe’s ingredients into a list showing comma separation
        recipe_ingredients_list = [
            ing.lower().strip() for ing in row["ingredients_name"].split(",")
        ]
        
        
        if all(
            any(u_ing in r_ing for r_ing in recipe_ingredients_list)
            for u_ing in user_ingredients_list
        ):
        
            #checking the user input indgredient match and giving the rest as extra indgredient.
            extra_count = len(recipe_ingredients_list) - len(user_ingredients_list)
            matches.append((row, extra_count))
    
    
    matches.sort(key=lambda x: x[1])
    
    # Return the top_n recipes
    return matches[:top_n]
```

```python
from datasets import Dataset

def map_dataset(batch):
    return tokenizer(
        batch["name"],
        truncation=True,
        max_length=128,
        return_overflowing_tokens=True
    )

#mapping the data.
dataset = Dataset.from_pandas(df)
dataset = dataset.map(
    map_dataset,
    batched=True,
    batch_size=8,
    remove_columns=list(df.columns)
)
dataset = dataset.remove_columns(["overflow_to_sample_mapping"])
dataset = dataset.train_test_split(test_size=0.2)
dataset
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
    )
data_collator
```

```python
from transformers import TrainingArguments, Trainer
#training the model with the dataset
training_args = TrainingArguments(
    output_dir="./output/model",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    weight_decay=0.01,
    num_train_epochs=10

)

trainer = Trainer(
    model=model,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    args=training_args,
    data_collator=data_collator

)

trainer.train()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import re
import pandas as pd

def get_best_matches(df, user_ingredients, top_n=3):
    """
    Returns the top_n recipes from df that contain ALL user_ingredients
    by checking if each user ingredient is a substring of at least onecc
    recipe ingredient. Sorts by fewest extra ingredients.
    """
    user_ingredients_list = [ing.lower().strip() for ing in user_ingredients]
    
    matches = []
    for _, row in df.iterrows():
        # Recipe’s ingredients into a list showing comma separation
        recipe_ingredients_list = [
            ing.lower().strip() for ing in row["ingredients_name"].split(",")
        ]
        
        # Checking if the inputted indgredient found in data
        if all(
            any(u_ing in r_ing for r_ing in recipe_ingredients_list)
            for u_ing in user_ingredients_list
        ):
           #Checking the user input indgredient match and giving the rest as extra indgredient.
            extra_count = len(recipe_ingredients_list) - len(user_ingredients_list)
            matches.append((row, extra_count))
    
    
    matches.sort(key=lambda x: x[1])
    
    #Return the top_n recipes
    return matches[:top_n]
# Model used for the traininig
model_name = "distilgpt2"
device = "cpu"



df = pd.read_csv("./Food_Recipe.csv")
df = df[:1500]  
df

tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained("./output/model/checkpoint-3190")

prompt = "a recipe using bread,butter,cheese,garlic,spinach"

# Generate text
inputs = tokenizer(prompt, return_tensors="pt").to(device)
outputs = model.generate(
    **inputs,
    max_new_tokens=100,
    do_sample=True,
    top_k=50,
    top_p=0.95
)
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("=== Model’s Generated Text ===")
print(generated_text)


pattern = r"using\s+(.*)"
match = re.search(pattern, prompt, flags=re.IGNORECASE)
if match:
    # Ingredient seperated by comma
    user_ingredients = [ing.strip() for ing in match.group(1).split(",")]
else:
    
    user_ingredients = []

# Get top 3 matches
best_matches = get_best_matches(df, user_ingredients, top_n=3)

# Print them out
print("\n=== Top 3 Matching Recipes ===")
for i, (row, extra_count) in enumerate(best_matches, start=1):
    print(f"\nMatch #{i}")
    print(f"Recipe Name: {row['name']}")
    print(f"Extra Ingredients Needed: {extra_count}")
    print(f"Instructions: {row['instructions']}")
```
